<h2> Images and NumPy</h2>

We first need to import the required libraries
matplotlib is a very usful data visualisation tool
The Image library is used to load images into python

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
```

The picture is loaded

```python
Picture=Image.open('../Data/Puppy.jpg')
```

```python
Picture
```

The picture is in the form of a JpegFile. We need to convert this JpegFile into a NumPy array.

```python
type(Picture)
```

```python
PictureArray=np.array(Picture)
```

```python
print(type(PictureArray))
```

Each pixel in a image is defined in terms of colour channels
In general, most images follow RGB colour channel configuration.
R-Red,G-Green,B-Blue
The mix of R,G,B creates all the colours visible in the image.

As seen below the shape of the array is (453,680,3)
453 is the height of the image
680 is the width of the image
3 denotes the 3 colour channels RGB

```python
print(PictureArray.shape)
```

```python
plt.imshow(PictureArray)
```

PictureArray[:,:,0]=R
PictureArray[:,:,1]=G
PictureArray[:,:,2]=B
The values of these channels are usually 0-255

Now let us check the intensity of the R channel in the image.
Let us create a copy of the array to pursue this further.

```python
PicRed=PictureArray.copy()
plt.imshow(PicRed[:,:,0])
```

The above image is supposed to show the intensity of the values of Red channel, however the image seems multicoloured and confusing. This is because the cmap parameter of the .imshow() has not been set. 

Let us set the cmap as gray.
This means that places of lesser R intensity are more whitish.
Places where intensity is 0 are pure white

Places of higher R intensity are more blackish.
Places where intensity is highest(255) are pure black.

Intermediate values are defined in shades of gray.

Now the intensity chart is displayed correctly.

```python
plt.imshow(PicRed[:,:,0],cmap='gray')
```

Let's also see the intensity of the G channel.

```python
plt.imshow(PicRed[:,:,1],cmap='gray')
```

Let's also see the intensity of the B channel.

```python
plt.imshow(PicRed[:,:,2],cmap='gray')
```

Lets set the intensity of the G and B channels as 0 and check the final picture

```python
PicRed[:,:,1]=0
PicRed[:,:,2]=0
```

```python
plt.imshow(PicRed)
```

```python
PicGreen=PictureArray.copy()
PicGreen[:,:,0]=0
PicGreen[:,:,2]=0
```

```python
plt.imshow(PicGreen)
```

```python
PicBlue=PictureArray.copy()
PicBlue[:,:,0]=0
PicBlue[:,:,1]=0
```

```python
plt.imshow(PicBlue)
```

Similarly, all images are made of a mixture of color channels.
Each indiviudal channel can be viewed by setting the intensity of the other channels to 0.
