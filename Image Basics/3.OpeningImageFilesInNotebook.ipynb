```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

<h2> Loading images </h2>

Images can be loaded in OpenCV using the .imread() method.
The relative path of the image must be given.

```python
Image=cv2.imread('Data/Puppy.jpg')
```

The image is stored as a NumPy array.

```python
type(Image)
```

Incase an invalid path is given, the .imread() returns a None value

```python
Image=cv2.imread('wrong/path')
print(Image)
print(type(None))
```

```python
Image=cv2.imread('Data/Puppy.jpg')
```

As discussed before, the Image has 3 colour channels and this can be verified using the .shape parameter

```python
print(Image.shape)
```

Let's try viewing the image using plt

```python
plt.imshow(Image)
```

The image is weird because of the different ways plt and OpenCV interpret the colour channels.


plt considers the order of color channels to be RGB but OpenCV stores colour data in BGR 

<h2>Converting colours</h2>

We can change the way OpenCV stores colour channels by using the .cvtcolor() function

```python
Image=cv2.cvtColor(Image,cv2.COLOR_BGR2RGB)
```

It will now display colours properly in PLT

```python
plt.imshow(Image)
```

Colour channels can also be specified during .imread()

```python
Image_Gray=cv2.imread('Data/Puppy.jpg',cv2.IMREAD_GRAYSCALE)
```

The above code will convert the image into grayscale during read

Grayscale images can display colours using a single colour channel (0-255) so the shape of the image is reduced.

```python
print(Image_Gray.shape)
```

```python
plt.imshow(Image_Gray)
```

The above image is weird because plt applies a colour map to 2d Arrays.

So we need to specify the colourmap as gray

```python
plt.imshow(Image_Gray,cmap='gray')
```

<h2>Resizing images</h2>

It's possible to resize the width and height of an image using the .resize() method

```python
Resized_Image=cv2.resize(Image,(100,100))
```

```python
plt.imshow(Resized_Image)
```

```python
print(Resized_Image.shape)
```

As seen the image is resized with no changes to the color

It is also possible to resize using ratios.When the second parameter is set as (0,0) it is possible to enter width and height ratio  in the further parametrs.


Parametes:
1.Original image<br /> 
2.(0,0)<br /> 
3.Destination Image (this parameters seems to be redundant so just give the original image here as well)<br /> 
4.X axis scale factor <br /> 
5.Y axis scale factor

```python
w_ratio=0.5
h_ratio=0.5
```

```python
Resized_Image=cv2.resize(Image,(0,0),Image,w_ratio,h_ratio)
```

```python
plt.imshow(Resized_Image)
```

```python
print(Resized_Image.shape)
```

<h2>Flipping images</h2>

Next, let's look at flipping images.<br /> 
Flipping images can be done using the cv2.flip() method<br /> 
cv2.flip(Image,0) flips the image along X Axis<br /> 
cv2.flip(Image,1) flips the image along Y Axis<br /> 
cv2.flip(Image,-1) flips the image along both X and Y Axis<br /> 

```python
FlippedImage=cv2.flip(Image,0)
```

```python
plt.imshow(FlippedImage)
```

```python
FlippedImage=cv2.flip(Image,1)
```

```python
plt.imshow(FlippedImage)
```

```python
FlippedImage=cv2.flip(Image,-1)
```

```python
plt.imshow(FlippedImage)
```
