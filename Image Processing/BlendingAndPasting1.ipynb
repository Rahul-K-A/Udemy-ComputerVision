<h2>Blending and Pasting Images</h2>

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img1=cv2.imread("../Data/Space.jpg")
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
print(img1.shape)
```

```python
img2=cv2.imread("../Data/StopSign.jpg")
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.imshow(img2)
print(img2.shape)
```

Images of the same size can be blended together by using the cv2.addWeighted() function

Add weighted works as follows: <br>
BlendedImage= alpha * Image1 + beta * Image2 + gamma

Alpha and beta are the weights of the images. The higher the weight, the more the image will be present in the blended image. Gamma basically manipulates the luminance of the image. High gamma values will turn the image white.

<a href="https://en.wikipedia.org/wiki/Gamma_correction">To learn more about the role of Gamma, click here</a>

```python
blended=cv2.addWeighted(src1=img1,alpha=0.75,src2=img2,beta=0.25,gamma=0)
plt.imshow(blended)
```

Overlaying two images which are not of the same size is also possible, though it has to be done manually and without any blending. i.e the images will be opaque.

```python
LargeImage=cv2.imread("../Data/Butterfly.jpg")
plt.imshow(LargeImage)
SmallImage=cv2.imread("../Data/Space.jpg")
plt.imshow(SmallImage)
```

```python
XOffset=0
YOffset=0
XEnd=XOffset+SmallImage.shape[1]
YEnd=YOffset+SmallImage.shape[0]
LargeImage[YOffset:YEnd,XOffset:XEnd]=SmallImage
plt.imshow(LargeImage)
```
