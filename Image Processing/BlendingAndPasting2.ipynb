<h1> Blending and pasting 2-Masks</h1>

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img1=cv2.imread("../Data/Butterfly.jpg")
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
print(img1.shape)
```

```python
img2=cv2.imread("../Data/Stop.png")
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.imshow(img2)
print(img2.shape)
```

```python
XOffset=600-300
YOffset=600-300
RegionOfInterest=img1[YOffset:600,XOffset:600]
plt.imshow(RegionOfInterest)
```

```python
img2gray=cv2.cvtColor(img2,cv2.COLOR_RGB2GRAY)
print(img2gray.shape)
plt.imshow(img2gray,cmap='gray')
```

```python
InvertedMask=cv2.bitwise_not(img2gray)
```

```python
Foreground=cv2.bitwise_or(img2,img2,mask=InvertedMask)
plt.imshow(Foreground)
```

```python
FinalImage=cv2.bitwise_or(RegionOfInterest,Foreground)
plt.imshow(FinalImage)
```

```python
img1[YOffset:600,XOffset:600]=FinalImage
plt.imshow(img1)
```
